```python
import pandas as pd
import altair as alt

# Load Iris dataset using CSV URL
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
iris = pd.read_csv(url)

# Create scatter plot
scatter_plot = alt.Chart(iris).mark_circle().encode(
    x='petal_length',
    y='petal_width',
    color='species'
).properties(
    width=400,
    height=400
)

# Show the plot
scatter_plot
```

```python
import pandas as pd
import pylab as plt
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

Iris = pd.read_csv('iris.csv')
# Made the figure relatively large for clarity purposes 
fig = plt.figure(figsize=(20, 10))

axis1 = plt.subplot(122)

parallel_coordinates(Iris, 'variety', color = ['b', 'y', 'r'])

axis1.spines["top"].set_visible(False) 
axis1.spines["bottom"].set_visible(False)
axis1.spines["left"].set_visible(False)

plt.ylabel('cm') #label measurements

plt.tight_layout()
plt.show()
```
